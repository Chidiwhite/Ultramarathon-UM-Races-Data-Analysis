### Project Title: Ultramarathon (UM) Races Data Analysis (2020)

## Tools & Libraries: Python, Pandas, NumPy, Matplotlib, Seaborn

# Overview:
This project analyzes ultramarathon race data from the Two Centuries of UM Races dataset. The focus was on U.S.-based races of 50km and 50mi distances held in 2020. Using Python’s data analytics libraries, the project explored athlete demographics, performance differences, and participation trends.

# Objectives:

Identify performance gaps between male and female athletes.

Determine the best-performing age groups in 50mi races.

Examine how age correlates with average speed.

Analyze participation patterns across race distances.

# Key Insights:

Noticeable average speed difference between genders across both race lengths.

Athletes within the mid-30s age bracket consistently performed strongest in 50mi races (minimum 20 participants).

Longer races tended to show reduced average speeds, indicating endurance impact.

Skills Demonstrated:
Data cleaning, transformation, and filtering with pandas; data visualization with Seaborn; exploratory data analysis (EDA); and analytical storytelling.

Outcome:
Delivered actionable insights into ultramarathon performance dynamics and built a clear, data-driven narrative suitable for research or sports analytics applications.
